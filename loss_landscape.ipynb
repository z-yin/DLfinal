```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.style as style
style.use('seaborn')
import numpy as np
import pickle
import os
```

```python
with open('gn_0.0001.pickle', 'rb') as f:
    gn_0001 = pickle.load(f)

with open('bn_0.0001.pickle', 'rb') as f:
    bn_0001 = pickle.load(f)
    
with open('std_0.0001.pickle', 'rb') as f:
    std_0001 = pickle.load(f)
    
    
with open('gn_0.0005.pickle', 'rb') as f:
    gn_0005 = pickle.load(f)

with open('bn_0.0005.pickle', 'rb') as f:
    bn_0005 = pickle.load(f)
    
with open('std_0.0005.pickle', 'rb') as f:
    std_0005 = pickle.load(f)
    
    
with open('gn_0.00075.pickle', 'rb') as f:
    gn_00075 = pickle.load(f)

with open('bn_0.00075.pickle', 'rb') as f:
    bn_00075 = pickle.load(f)
    
with open('std_0.00075.pickle', 'rb') as f:
    std_00075 = pickle.load(f)
    
    
with open('gn_0.001.pickle', 'rb') as f:
    gn_001 = pickle.load(f)

with open('bn_0.001.pickle', 'rb') as f:
    bn_001 = pickle.load(f)
    
with open('std_0.001.pickle', 'rb') as f:
    std_001 = pickle.load(f)
    

with open('gn_0.005.pickle', 'rb') as f:
    gn_005 = pickle.load(f)

with open('bn_0.005.pickle', 'rb') as f:
    bn_005 = pickle.load(f)
    
with open('std_0.005.pickle', 'rb') as f:
    std_005 = pickle.load(f)
```

```python
gn_curve_min = []
gn_curve_max = []
for i in range(len(gn_0001)):
    gn_curve_min.append(np.min([gn_0005[i], gn_00075[i], gn_001[i]]))
    gn_curve_max.append(np.max([gn_0005[i], gn_00075[i], gn_001[i]]))

bn_curve_min = []
bn_curve_max = []
for i in range(len(bn_0001)):
    bn_curve_min.append(np.min([bn_0005[i], bn_00075[i], bn_001[i]]))
    bn_curve_max.append(np.max([bn_0005[i], bn_00075[i], bn_001[i]]))
    
std_curve_min = []
std_curve_max = []
for i in range(len(std_0001)):
    std_curve_min.append(np.min([std_0005[i], std_00075[i], std_001[i]]))
    std_curve_max.append(np.max([std_0005[i], std_00075[i], std_001[i]]))
```

```python
step = 30
steps = np.arange(0, len(gn_curve_min), step)
plt.figure(figsize=(15, 10))
plt.fill_between(steps, gn_curve_min[::step], gn_curve_max[::step],
                alpha=0.5, color='C1', label='Standard + GN')
plt.plot(steps, gn_curve_min[::step], color='C1')
plt.plot(steps, gn_curve_max[::step], color='C1')

plt.fill_between(steps, bn_curve_min[::step], bn_curve_max[::step],
                alpha=0.5, color='C2', label='Standard + BN')
plt.plot(steps, bn_curve_min[::step], color='C2')
plt.plot(steps, bn_curve_max[::step], color='C2')

plt.fill_between(steps, std_curve_min[::step], std_curve_max[::step],
                alpha=0.5, color='C3', label='Standard')
plt.plot(steps, std_curve_min[::step], color='C3')
plt.plot(steps, std_curve_max[::step], color='C3')

plt.legend(fontsize=19)
plt.title('Loss Landscape', fontsize=20)
plt.ylabel('Loss Landscape', fontsize=13)
plt.xlabel('Steps', fontsize=13)
plt.savefig(os.path.join('figs', 'loss_landscape.png'), dpi=500, quality=100)
plt.show()
```
